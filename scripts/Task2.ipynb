```python
def data_preprocessing(data, masterclass, drop_col):
    frame = data.dropna()
    time_col  = pd.to_datetime(data['Timestamp']).apply(lambda x: x.time())
    date_col = pd.to_datetime(data['Timestamp']).apply(lambda x: x.date())
    data["date"] = date_col
    data["time"] = time_col
    data["masterclass"] = masterclass
    data.drop(drop_col, axis=1, inplace=True)
    return data
    
    
```

```python
Masterclass = VE.append(CAN, ignore_index=True).append(DM, ignore_index=True).append(EXL, ignore_index=True).append(POD, ignore_index=True).append(PUB, ignore_index=True).append(WM, ignore_index=True)
```

```python
Masterclass['City'] = Masterclass['City'].str.lower()
Masterclass['date'] = pd.to_datetime(Masterclass['date'])

Masterclass['time'] = Masterclass['time'].astype(str)
```

```python
time_ = []
for i in Masterclass['time']:
    time_.append(i.split(".")[0])

Masterclass['time'] = time_
Masterclass['hour'] = pd.to_datetime(Masterclass['time'], format='%H:%M:%S').dt.hour
Masterclass['time'] = pd.to_datetime(Masterclass['time'], format='%H:%M:%S').dt.time[0]
Masterclass
```
